```python
"""
This is for Windows only
- go to https://pandoc.org/ to get pandoc first (Windows/Linux/MacOS)
- install MS word on machine
"""

import pypandoc
print(pypandoc.get_pandoc_version())
print(pypandoc.get_pandoc_path())
print(pypandoc.get_pandoc_formats())

import re, os, time
import win32com.client as win32
from win32com.client import constants
from glob import glob
```

```python
p = os.path.join(os.getcwd(),"**\\*")
```

```python
# recursively search the current working folders and subfolers
paths = glob(os.path.join(os.getcwd(),"**\\*"), recursive=True)
```

```python
for path in paths:        
    # Rename path with .md
    print("current:", path)

    if "DistrictCourt" not in path: continue

    if path.endswith('.doca'):
        print("converting to docx")

        if (os.path.isfile(path.replace(u'doc', u'txt'))) == True:
            continue

        # Opening MS Word
        word = win32.gencache.EnsureDispatch('Word.Application')
        doc = word.Documents.Open(path)
        doc.Activate()

        # Rename path with .docx
        new_file_abs = path.replace(u'doc', u'docx')

        # Save and Close
        word.ActiveDocument.SaveAs(
            new_file_abs, FileFormat=constants.wdFormatXMLDocument
        )
        doc.Close(False)
        
        # convert to txt using pypandoc
        print("converting to md")
        newfile = path.replace(u'doc', u'txt')
        pypandoc.convert_file(new_file_abs, 'plain', outputfile=newfile)

    elif path.endswith('.docx'):
        if (os.path.isfile(path.replace(u'docx', u'txt'))) == True:
            continue

        # convert to md using pypandoc
        print("converting to txt")
        newfile = path.replace(u'docx', u'txt')
        pypandoc.convert_file(path, 'plain', outputfile=newfile)
    
    time.sleep(0)

print("finished")
```
